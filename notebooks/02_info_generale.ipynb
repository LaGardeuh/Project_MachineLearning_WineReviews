```python
import pandas as pd
from pandas import isnull

file_path = "../data/cleaned/wine_data_cleaned.csv"

df_clean = pd.read_csv(file_path)

print("=" * 50)
print("SHAPE (dimensions du dataset)")
print("=" * 50)
print(f"Nombre de lignes : {df_clean.shape[0]}")
print(f"Nombre de colonnes : {df_clean.shape[1]}")
print(f"Dimensions : {df_clean.shape}")

print("\n" + "=" * 50)
print("COLUMNS (liste des colonnes)")
print("=" * 50)
print(df_clean.columns.tolist())

print("\n" + "=" * 50)
print("INFO (types de données et valeurs non-nulles)")
print("=" * 50)
df_clean.info()

print("\n" + "=" * 50)
print("DESCRIBE (statistiques descriptives)")
print("=" * 50)
print(df_clean.describe())

# Pour inclure aussi les colonnes non-numériques
print("\n" + "=" * 50)
print("DESCRIBE - TOUTES LES COLONNES")
print("=" * 50)
print(df_clean.describe(include='all'))
```
